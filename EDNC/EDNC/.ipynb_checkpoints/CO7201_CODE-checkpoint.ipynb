# Train, validation, test split

```python
covid_train = np.random.choice(COVID_files, size=739, replace=False) #non repeat randomly choose covid imgs
non_covid_train = np.random.choice(non_COVID_files, size=739, replace=False)
covid_files = list(set(COVID_files) - set(covid_train)) #subtract used training dataset
non_covid_files = list(set(non_COVID_files) - set(non_covid_train))

covid_val = np.random.choice(covid_files, size=245, replace=False)
non_covid_val = np.random.choice(non_COVID_files, size=245, replace=False)
covid_files = list(set(covid_files) - set(covid_val))
non_covid_files = list(set(non_covid_files) - set(non_covid_val))

covid_test = np.random.choice(covid_files, size=245, replace=False)
non_covid_test = np.random.choice(non_covid_files, size=245, replace=False)

print('COVID datasets:', covid_train.shape, covid_val.shape, covid_test.shape)
print('non-COVID datasets:', non_covid_train.shape, non_covid_val.shape, non_covid_test.shape)
```

```python
train_dir = 'training_data' #add a directory
val_dir = 'validation_data'
test_dir = 'test_data'

train_files = np.concatenate([covid_train, non_covid_train])
validate_files = np.concatenate([covid_val, non_covid_val])
test_files = np.concatenate([covid_test, non_covid_test])

if not os.path.isdir(train_dir):
    os.mkdir(train_dir) #create a directory named 'training_data'
else: 
    None 

if not os.path.isdir(val_dir):
    os.mkdir(val_dir) 
else:
    None
    
if not os.path.isdir(test_dir):
    os.mkdir(test_dir) 
else:
    None

for file in train_files:
    shutil.copy(file, train_dir) # copy the content of train_files to train_dir

for file in validate_files:
    shutil.copy(file, val_dir)
    
for file in test_files:
    shutil.copy(file, test_dir)
```

# Image to array

```python
IMG_DIM = (224, 224)

train_files = glob.glob('training_data/*')
```

```python

#train_imgs=[]
for img in train_files:
    train_img = [img_to_array(load_img(img, target_size=IMG_DIM)) ]
    train_imgs=train_imgs+train_img
train_imgs = np.array(train_imgs) 
```

```python
train_imgs = [img_to_array(load_img(img, target_size=IMG_DIM)) for img in train_files]
train_imgs = np.array(train_imgs)
```

```python
validation_files = glob.glob('validation_data/*')
#validation_imgs=[]
for img in validation_files:
    validation_img = [img_to_array(load_img(img, target_size=IMG_DIM))]
    validation_imgs=validation_img+ validation_imgs  
validation_imgs = np.array(validation_imgs)   
```

```python
test_files = glob.glob('test_data/*')
#test_imgs=[]
for img in test_files:
    test_img = [img_to_array(load_img(img, target_size=IMG_DIM))]
    test_imgs=test_img+test_imgs
test_imgs = np.array(test_imgs) 
```

# Retrive labels

```python
train_labels = [fn.split('/')[1].split()[0].strip() for fn in train_files]
train_labels[0]

validation_labels = [fn.split('/')[1].split()[0].strip() for fn in validation_files]
validation_labels[6]

test_labels = [fn.split('/')[1].split()[0].strip() for fn in test_files]
test_labels[0]
```

# Image normalization

```python
train_imgs_scaled = train_imgs.astype('float32')
train_imgs_scaled /= 255

validation_imgs_scaled  = validation_imgs.astype('float32')
validation_imgs_scaled /= 255

test_imgs_scaled = test_imgs.astype('float32')
test_imgs_scaled /= 255

print(train_imgs_scaled[0].shape)
array_to_img(train_imgs_scaled[0])
```

# Label encoding

```python
from sklearn.preprocessing import LabelEncoder #replace the categorical value with a numeric value
le = LabelEncoder()
train_labels_enc = le.fit_transform(train_labels)
print(train_labels[14:20], train_labels_enc[14:20])
validation_labels_enc = le.fit_transform(validation_labels)
print(validation_labels[14:20], validation_labels_enc[14:20])
test_labels_enc = le.fit_transform(test_labels)
print(test_labels[14:20], test_labels_enc[14:20])
```

# One hot encoding

```python
from keras import utils as np_utils 
train_one_hot_label = tf.keras.utils.to_categorical(train_labels_enc)
val_one_hot_label = tf.keras.utils.to_categorical(validation_labels_enc)
test_one_hot_label = tf.keras.utils.to_categorical(test_labels_enc)
```

```python
print(val_one_hot_label[14:20])
```

# Image augementation

```python
image_size = 224 
BATCH_SIZE = 16 
STEPS_PER_EPOCH = len(train_imgs)/BATCH_SIZE
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   zoom_range = 0.2,
                                   rotation_range=15,
                                   horizontal_flip = True)

val_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
train_generator = train_datagen.flow(train_imgs, train_one_hot_label, batch_size= BATCH_SIZE )
val_generator = val_datagen.flow(validation_imgs, val_one_hot_label, batch_size= BATCH_SIZE )
```

```python
lr_reduce = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, epsilon=0.0001, patience=3, verbose=1)
es_callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, verbose=1)
```

# Densenet201

```python
pretrained_densenet = tf.keras.applications.DenseNet201(input_shape=(image_size, image_size, 3), weights='imagenet', include_top=False)
```

```python

for layer in pretrained_densenet.layers:
  layer.trainable = False

x1 = pretrained_densenet.output
x1 = tf.keras.layers.AveragePooling2D(name="averagepooling2d")(x1)
x1 = tf.keras.layers.Flatten(name="flatten")(x1)
x1 = tf.keras.layers.Dense(64, activation="relu", name="dense")(x1)
x1 = tf.keras.layers.Dropout(0.5, name="dropout")(x1)
model_out = tf.keras.layers.Dense(2, activation='softmax', name="predictions")(x1)

model_densenet = Model(inputs=pretrained_densenet.input, outputs=model_out)
model_densenet.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),loss='categorical_crossentropy',metrics=['accuracy'])

model_densenet.summary()
```

```python
history_densenet = model_densenet.fit_generator(train_generator, validation_data=val_generator, callbacks=[lr_reduce, es_callback], epochs=20)
```

```python
# accuracy plot 
plt.plot(history_densenet.history['accuracy'])
plt.plot(history_densenet.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# loss plot
plt.plot(history_densenet.history['loss'])
plt.plot(history_densenet.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
Y_pred = model_densenet.predict(test_imgs_scaled)
Y_pred = np.argmax(Y_pred, axis=1)
Y_true = np.argmax(test_one_hot_label, axis=1)
Y_true.shape
from sklearn.metrics import confusion_matrix
from sklearn import metrics
import seaborn as sns
cm = confusion_matrix(Y_true, Y_pred)
print(metrics.confusion_matrix(Y_true, Y_pred))
# Printing the precision and recall, among other metrics
print(metrics.classification_report(Y_true, Y_pred))
```

```python
disease_types=['COVID', 'non-COVID']
plt.figure(figsize=(7, 7))
ax = sns.heatmap(cm, cmap=plt.cm.Blues, annot=True, square=True, xticklabels=disease_types, yticklabels=disease_types)
plt.xlabel('Predicted',fontsize=20)
plt.ylabel('True',fontsize=20)
```

```python
model_densenet.save('dense.h5')
```

# Customized CNN

```python
input_shape = (224,224, 3)
from tensorflow.keras import optimizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, BatchNormalization, AveragePooling2D, GlobalAveragePooling2D
```

```python
model = Sequential()

model.add(Conv2D(16, kernel_size=(3, 3), activation='relu', 
                 input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(2, activation='softmax'))


model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.RMSprop(),
              metrics=['accuracy'])

model.summary()
```

```python
history= model.fit(train_generator, validation_data=val_generator, callbacks=[lr_reduce, es_callback], epochs=50)
```

```python
# accuracy plot 
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# loss plot
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
Y_pred = model(test_imgs_scaled)
Y_pred = np.argmax(Y_pred, axis=1)
Y_true = np.argmax(test_one_hot_label, axis=1)

cm = confusion_matrix(Y_true, Y_pred)
print(metrics.confusion_matrix(Y_true, Y_pred))
# Printing the precision and recall, among other metrics
print(metrics.classification_report(Y_true, Y_pred))
```

```python
disease_types=['COVID', 'non-COVID']
plt.figure(figsize=(7, 7))
ax = sns.heatmap(cm, cmap=plt.cm.Blues, annot=True, square=True, xticklabels=disease_types, yticklabels=disease_types)
plt.xlabel('Predicted',fontsize=20)
plt.ylabel('True',fontsize=20)
```
